#Requirements
- An XLSX file exported from the Not Found(404) Index Coverage Report on your Google Search Console property
- An XLSX file exported from the Ahrefs' Backlinks report 


> Please, refer to the [full guide](https://seodepths.com/python-for-seo/broken-backlinks-audit-pyhton/) to make it through the first steps

```python
#@title Import Pages with Errors from GSC

#Please make sure to remove spaces in the above rows containing labels as they may not be readable

import pandas as pd
gsc = pd.read_excel('/content/vault gsc 404.xlsx')
df = pd.DataFrame(gsc, columns=['URL','Last crawled']) 
df.head()
```

```python
#@title Import the Ahrefs Backlinks report
internal = pd.read_excel('/content/vault backlinks.xlsx')
df2 = pd.DataFrame(internal, columns=['URL','External links'])	
df2.head()
```

```python
#@title Operate the Vlookup with Pandas
backlinks = df.merge(df2,  on='URL')
backlinks
```

```python
#@title Data Cleaning - Remove Duplicate
output = backlinks.drop_duplicates('URL')
output
```

```python
#@title Print 404 pages with backlinks

#sort external links by descending order
x = output.sort_values(['External links'], ascending=False).reindex(columns=['URL','External links','Last crawled'])
#save to excel
x.to_excel('2022122_Gucci_Vault_Historic_Broken_Links.xlsx', index=False)
#style the output
x.style.background_gradient()
```
